# 7. Store the target column (i.e.Personal Loan) in the y variable and the rest of the columns in the Xvariable.

```python

# store the target column in a variable y
y = bank['Personal Loan']

# store the rest of the columns in a variable X
X = bank.drop('Personal Loan', axis=1)
```

# 8. Split the dataset into two parts (i.e. 70% train and 30% test). and standardize the columns using the z-score scaling approach.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split the dataset into 70% train and 30% test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the columns of the train and test sets using z-score scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

#### We separate the target variable from the features by dropping the 'Personal Loan' column from the features and storing it in a separate variable named 'y.' --> Next, we use scikit-learn's train_test_split function to split the dataset into 70% train and 30% test sets. We set the test_size parameter to 0.3 to indicate a 30% test set size, and we set the random_state parameter to 42 to ensure reproducibility of the results. --> Finally, we use scikit-learn's StandardScaler to standardize the columns of the train and test sets using z-score scaling. We fit the scaler on the train set and transform both the train and test sets using the fitted scaler. The standardized train and test sets are stored in variables named X_train_scaled and X_test_scaled, respectively.

# 9. Train and test a Logistic Regression model to predict the likelihood of a liability customer buying personal loans. Display the train and test accuracy scores.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Create a Logistic Regression model object
model = LogisticRegression()

# Train the model on the standardized train set
model.fit(X_train_scaled, y_train)

# Predict the train and test sets using the trained model
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

# Compute the accuracy scores of the train and test sets
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# Print the accuracy scores of the train and test sets
print('Train accuracy:', train_accuracy)
print('Test accuracy:', test_accuracy)
```

#### We then create a Logistic Regression model object named "model." --> Next, we train the model on the standardized train set using the model.fit() method. --> We then use the trained model to predict the train and test sets using the model.predict() method, and store the predicted values in variables named y_train_pred and y_test_pred, respectively. --> Finally, we compute the accuracy scores of the train and test sets using scikit-learn's accuracy_score() method, and store the scores in variables named train_accuracy and test_accuracy, respectively. We print the train and test accuracy scores using the print() function.

# 10. Print the confusion matrix and classification report for the model and write your conclusions on the results.

```python
from sklearn.metrics import confusion_matrix, classification_report

# Compute the confusion matrix and classification report of the test set
confusion_mat = confusion_matrix(y_test, y_test_pred)
classification_rep = classification_report(y_test, y_test_pred)

# Print the confusion matrix and classification report of the test set
print('Confusion matrix:\n', confusion_mat)
print('Classification report:\n', classification_rep)
```
